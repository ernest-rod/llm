```python
from binary_file_reader import (
    BinaryFileReader, 
    RecordFormat,
    ErrorHandlingMode
)
```

```python
# 1. Define your binary structure (matching C struct)
record_format = RecordFormat(
    format_string='i50s50s100s20s50s3s10s11s',  
    byte_order='<'           # Little-endian
)
```

```python
# 2. Create reader
reader = BinaryFileReader(record_format=record_format)
```

```python
# 3. Read the file
result = reader.read_file('data\\customers.binary') 
```

```python
# 4. Use the results
result.print_summary()
```

```python
for record in result.records:
    print(f"ID: {record['id']}, First Name: {record['first_name'].decode().strip()}, Last Name: {record['last_name'].decode().strip()}, Email: {record['email'].decode().strip()}")
```
