```python
from binary_file_reader_flexible import (
    BinaryFileReader,
    RecordSchema,
    FieldDefinition,
    FieldType
)
```

```python
# Define your binary structure
schema = RecordSchema(
    fields=[
        FieldDefinition(name='customer_id', field_type=FieldType.INT32),
        FieldDefinition(name='first_name', field_type=FieldType.STRING, string_length=50),
        FieldDefinition(name='last_name', field_type=FieldType.STRING, string_length=50),
        FieldDefinition(name='email', field_type=FieldType.STRING, string_length=100),
        FieldDefinition(name='phone', field_type=FieldType.STRING, string_length=20),
        FieldDefinition(name='city', field_type=FieldType.STRING, string_length=50),
        FieldDefinition(name='state', field_type=FieldType.STRING, string_length=3),
        FieldDefinition(name='zip_code', field_type=FieldType.STRING, string_length=10),
        FieldDefinition(name='registration_date', field_type=FieldType.STRING, string_length=11)
    ],
    byte_order='<',  # Little-endian
    name="My Data Structure"
)
```

```python
# Create reader and read file
reader = BinaryFileReader(schema=schema)
result = reader.read_file('data\\customers.binary')
```

```python
# Print summary and access records
result.print_summary()
for record in result.records:
    print(record)
```
